## Top 10 Videos by Views

```python
import requests
import pandas as pd

# Get Your API from a json file
import json
with open(r"C:\Users\orena\OneDrive\01_ORENAIKE\02_CAREER_AND_DEVELOPMENTS\01_Schools\ALX\Codes\APIs\Api.json") as f:
  data = json.load(f)
  api_key = data["YoutubeApi_Key"]
  
# Channel IDs
channel_ids = ['UCW8Ews7tdKKkBT6GdtQaXvQ']

# Initialize an empty list to store temporary DataFrames
temp_dfs = []

# API Key and Endpoint
#api_key = 'YOUR_API_KEY'
endpoint = "https://www.googleapis.com/youtube/v3/search"

# Parameters for the API request
params = {
    "part": "snippet",
    "chart": "mostPopular",
    # "regionCode": "US",
    "key": api_key,
    "channelId": 'UCoOae5nYA7VqaXzerajD0lg',
    "maxResults": 1000,
    # "order": "date"
    "order" : "viewCount"
# UCoOae5nYA7VqaXzerajD0lg
# UCW8Ews7tdKKkBT6GdtQaXvQ

}

# UCoOae5nYA7VqaXzerajD0lg
# Make the API request
response = requests.get(endpoint, params=params).json()

# Loop through each video
for video in response['items']:
    if video['id']['kind'] == 'youtube#video':
        video_id = video['id']['videoId']
        video_title = video['snippet']['title']
        publishDate = video['snippet']['publishedAt'].split("T")[0]

        # API request for video statistics
        stats_endpoint = "https://www.googleapis.com/youtube/v3/videos"
        stats_params = {
            "part": "statistics",
            "key": api_key,
            "id": video_id
        }

        stats_response = requests.get(stats_endpoint, params=stats_params).json()
        viewCount = stats_response['items'][0]['statistics']['viewCount']
        likeCount = stats_response['items'][0]['statistics']['likeCount']
        commentCount = stats_response['items'][0]['statistics']['commentCount']

        # Create a temporary DataFrame for this video
        temp_df = pd.DataFrame([{
            "videoId": video_id,
            "video_title": video_title,
            "publishDate": publishDate,
            "viewCount": viewCount,
            "likeCount": likeCount,
            "commentCount": commentCount
        }])

        # Append the temporary DataFrame to the list
        temp_dfs.append(temp_df)

# Concatenate all the temporary DataFrames
df = pd.concat(temp_dfs, ignore_index=True)

# Display the DataFrame
df.head(10)
```

```python

print(df.dtypes)

# Convert specific columns to integer
df['likeCount'] = df['likeCount'].astype(int)
df['viewCount'] = df['viewCount'].astype(int)
df['commentCount'] = df['commentCount'].astype(int)

# Sort the DataFrame by 'viewCount'
dfSorted = df.sort_values(by='viewCount', ascending=False)
dfSorted10 = dfSorted.head(10)
dfSorted10

# df_sorted = df.sort_values(by='likeCount', ascending=False)
```

###  Horizontal bar

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# df = pd.DataFrame(data)

# Sort DataFrame by likeCount in descending order
df_sorted = dfSorted10.sort_values(by='likeCount', ascending=False)

# Extract sorted data
labels_sorted = df_sorted['video_title']
likeCount_sorted = df_sorted['likeCount']
commentCount_sorted = df_sorted['commentCount']
x = np.arange(len(labels_sorted))

# Bar width
width = 0.2

# Create horizontal bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(x, likeCount_sorted, width, label='Like Count')
ax.barh(x + width, commentCount_sorted, width, label='Comment Count')

# Set chart title and labels
ax.set_title('YouTube Video Metrics Sorted by Like Count')
ax.set_xlabel('Counts')
ax.set_yticks(x)
ax.set_yticklabels(labels_sorted)

# Set legend
ax.legend()

# Adjust layout
plt.tight_layout()

# Show the chart
plt.show()
```

### Scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np


# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(dfSorted10['likeCount'], dfSorted10['commentCount'])

# Add titles and labels
plt.title('Correlation between Like Count and Comment Count')
plt.xlabel('Like Count')
plt.ylabel('Comment Count')

# Add trend line
z = np.polyfit(dfSorted10['likeCount'], dfSorted10['commentCount'], 1)
p = np.poly1d(z)
plt.plot(dfSorted10['likeCount'], p(dfSorted10['likeCount']), "r--")

# Show plot
plt.show()
```
