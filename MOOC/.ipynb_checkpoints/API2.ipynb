## Write a function to get trending videos

```python
import requests
import json

# Function to get trending videos
def get_trending_videos(api_key, region_code='US', max_results=50):
    endpoint = "https://www.googleapis.com/youtube/v3/videos"
    
    params = {
        "part": "snippet,statistics",  # Get snippet and statistics data
        "chart": "mostPopular",        # Fetch most popular videos
        "regionCode": region_code,     # Region code, e.g., US
        "key": api_key,
        "maxResults": max_results      # Maximum results per request
    }

    response = requests.get(endpoint, params=params)

    # Check if the request was successful
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        return None
    
# Fetch trending videos
trending_videos = get_trending_videos(api_key)
trending_videos
```

```python
import requests
import pandas as pd

# API Key and endpoint
api_key = api_key
endpoint = "https://www.googleapis.com/youtube/v3/videos"
channel_id = ['UCoOae5nYA7VqaXzerajD0lg']

# Parameters for the API request to get trending videos
params = {
    "part": "snippet,statistics",
    "chart": "mostPopular",
    "regionCode": "US",
    "key": api_key,
    "maxResults": 50,
    "channelId": channel_id,
}

# Fetch trending videos
response = requests.get(endpoint, params=params).json()

# Create an empty DataFrame
df = pd.DataFrame(columns=["videoId", "title", "publishDate", "viewCount", "likeCount", "commentCount"])

# Process each video
for video in response['items']:
    video_id = video['id']
    title = video['snippet']['title']
    publish_date = video['snippet']['publishedAt'].split("T")[0]
    view_count = video['statistics'].get('viewCount')
    like_count = video['statistics'].get('likeCount')
    comment_count = video['statistics'].get('commentCount')

    # Append to DataFrame
    data = []
    for video in response['items']:
        data.append({
            "videoId": video_id,
            "title": title,
            "publishDate": publish_date,
            "viewCount": view_count,
            "likeCount": like_count,
            "channelId": channel_id,
            "commentCount": comment_count
        })

# Create DataFrame from the accumulated data
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```
