#### **Variable Types and scope, Closure and Prototype**
**1. Variable Type and Scope:**
Variable type refers to the data type of a variable, such as string, number, or boolean. In JavaScript, variables can be declared using the keywords "var", "let", or "const". The scope of a variable refers to the part of the code where the variable is accessible. In JavaScript, variables can have global scope (accessible throughout the code) or local scope (accessible only within a specific function or block of code). Understanding variable type and scope is important for writing efficient and effective JavaScript code. [practical file](./JavaFiles/Scope.js)

**2. Closure:**
A closure is a feature in JavaScript that allows a function to access variables from its outer scope, even after the outer function has returned. Closures are created when a function is defined inside another function and the inner function "closes over" variables from the outer function. This can be useful for creating private variables and functions in JavaScript. [practical file](./JavaFiles/Closure.js)

**3. Prototype:**
In JavaScript, every object has a prototype, which is a blueprint for creating new objects of the same type. The prototype defines the properties and methods that are available to objects of that type. When you create a new object using the "new" keyword, JavaScript creates a new object based on the prototype of the constructor function. This can be useful for creating objects that share common properties and methods. [practical file](./JavaFiles/Prototype.js)

#### **Exporting and importing modules in JS**
- https://javascript.info/modules-intro
- https://www.w3schools.com/js/js_modules.asp

In JavaScript, you can export values or functions from a module using two different mechanisms: Named Exports and Default Exports.

**Named Exports:**

You can create a names export individually or all at once. Named Exports allow you to export multiple values or functions from a module using their respective names. These exports are defined using the `export` keyword, followed by the name of the function or value being exported:
```
//person.js
export const name = "Jesse";
export const age = 40;
```

or all at once
```
//person.js
const name = "Jesse";
const age = 40;

export {name, age};
```
**EXPORTS**
**Default Exports:**
In defauls you can only have one default export. Default Exports allow you to export a single value or function from a module without specifying its name. When a module is imported, the default export can be assigned to any variable name you choose. These exports are defined using the `export default` syntax, followed by the value or function being exported:

```
//person.js
const message = () => {
const name = "Jesse";
const age = 40;
return name + ' is ' + age + 'years old.';
};

export default message;
```
or 
```
//person.js
export default const message = () => {
const name = "Jesse";
const age = 40;
return name + ' is ' + age + 'years old.';
};

```
**IMPORTS**
You can import modules into a file in two ways, based on if they are named exports or default exports.

Imports from named exports
```
import { name, age } from "./person.js";
```
Imports from deafault exports
```
import message from "./message.js";
```

#### **Difference between `modules.export` and `export default`**
`module.exports` and `export default` are both mechanisms for exporting values or functions from a module in JavaScript, but they are used in different module systems and are not interchangeable.

**module.exports** is used in CommonJS, which is the module system used in Node.js. In this system, you assign a value to module.exports to export it from a module. For example, the following code exports a Rectangle class using module.exports:

```
class Rectangle {
  constructor(width, height) {
    this.width = width;
    this.height = height;
  }

  get area() {
    return this.width * this.height;
  }
}

module.exports = Rectangle;

```
On the other hand, `export default` is used in ECMAScript modules (ESM), which is a standardized module system used in modern web browsers and supported by many tools such as webpack, Rollup, and Babel. In this system, you can export a value or function as the `default export` of a module using the export default syntax. For example, the following code exports a `Rectangle` class as the default export of a module:

```
export default class Rectangle {
  constructor(width, height) {
    this.width = width;
    this.height = height;
  }

  get area() {
    return this.width * this.height;
  }
}

```
When you import a module that uses `module.exports`, you use `require()` to load the module and get the exported value. When you import a module that uses `export default`, you use `import` to load the module and get the default exported value. Here are examples of importing modules that use both types of exports:
```
// Importing a module that uses module.exports
const Rectangle = require('./Rectangle');

const r = new Rectangle(5, 10);
console.log(r.area);  // Output: 50

// Importing a module that uses export default
import Rectangle from './Rectangle';

const r = new Rectangle(5, 10);
console.log(r.area);  // Output: 50

```
Note that in ECMAScript modules, you can also use Named Exports along with Default Exports, whereas in CommonJS you can only use Named Exports.

#### How to read and write a file using `fs` module
- [How To Work with Files using the fs Module in Node.js](https://www.digitalocean.com/community/tutorials/how-to-work-with-files-using-the-fs-module-in-node-js)

The `fs` module in Node.js provides a set of functions for working with the file system. You can use this module to read and write files in JavaScript. you can read a file async and synchronously.

There are two ways of reading files, **Promise-based syntax**, while the second one is **traditional callback-based syntax**. **Promise-based syntax** provides a more organized and readable way to handle asynchronous code execution, while **traditional callback-based syntax** can become messy and hard to read when dealing with complex asynchronous operations.
[Practical Examples of Reading and writing files in JS](./JavaFiles/ReadWriteFile.js)

```
const request = require('request');

request('https://example.com', function (error, response, body) {
  if (error) {
    console.error('Error:', error);
  } else {
    console.log('Response:', response);
    console.log('Body:', body);
  }
});

```

In the exampl above, `request` module was required and the `request` function was used to make GET request to `https://example.com`. The callback function has three paameter. 

- **`error`**: If an error occurs during the request, this parameter will contain the error object. Otherwise, it will be **`null`**.
- **`response`**: This parameter contains the HTTP response object. It includes the status code, headers, and other information about the response.
- **`body`**: This parameter contains the response body as a string.

In the callback function, we first check if an error occurred during the request. If there is an error, we log it to the console. If there is no error, we log the response object and the response body to the console.

Note that there are many options you can pass to the request function to customize the HTTP request, such as setting headers, setting the HTTP method, and passing data in the request body.

### ?How to select HTML elements in JavaScript
https://www.youtube.com/watch?v=YaptCAJFR5k&list=PLGevxjyhYfA6GOvbGgcfEcyC4aUlJJcPb&index=55

JavaScript provides six methods to select an element from the document.

- `getElementById` – search element by element_id
  ```
  let myElement = document.getElementById("myId");
  ```
  
- `getElementsByTagName` – search element by tag name (e.g., span, div)
  ```
  let myElements = document.getElementsByClassName("myClass");
  ```
- `getElementsByClassName` – search element by class name
  ```
  let myElements = document.getElementsByClassName("myClass");
  ```
- `getElementsByName` – search element by name attribute
  ```
  let myElements = document.getElementsByName("myName");
  ```
- `querySelector` – returns the first element that matches the specified selector
  ```
  let myElement = document.querySelector(".myClass");
  ```
- `querySelectorAll` – returns elements that match the specified selector
  ```
  let myElements = document.querySelectorAll("div.myClass");
  ```
