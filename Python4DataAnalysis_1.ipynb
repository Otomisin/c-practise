str.extract()

```python
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'col1': ['A12', 'B23', 'C34', 'D45'],
                   'col2': ['apple', 'banana', 'cherry', 'date']})

# Extract all digits from the 'col1' column
df['col1_digits'] = df['col1'].str.extract(r'(\\d+)')

# Extract the first three characters from the 'col2' column
df['col2_first_three'] = df['col2'].str.extract(r'^(\\w{3})')

# Extract all the last 2 digits characters from 'col1'
df['firstC'] = df['col1'].str.extract(r'^(\\w{1})')

print(df)
```

str.split()

The str.split() method in pandas is used to split a string into separate substrings based on a specified delimiter. Here's an example of how you can use it:

# Merging dataframe

There are several ways to merge dataframes in pandas, including concat, merge, and join. The choice of method will depend on the specific requirements of your task.

Concat: 

This function is used to stack dataframes vertically (i.e., by adding rows) or horizontally (i.e., by adding columns). For example:

```python
import pandas as pd

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

#Concatinate
df_vertical = pd.concat([df1, df2])
df_vertical
```

```python
pd.concat([df1, df2], axis= 1)
```

Merge: 

This function is used to merge dataframes based on a common column or index. For example:

```python
df1a = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

df2a = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']})

df_merged = pd.merge(df1a, df2a, on='key')
df_merged
# df1a
# df2a
```

join: 

This function is similar to merge, but it is specifically designed to merge dataframes based on their indexes. For example:

```python
df1b = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']},
                   index=['K0', 'K1', 'K2', 'K3'])

df2b = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=['K0', 'K1', 'K2', 'K3'])

df_joined = df1b.join(df2b)
df_joined
```

Function for date cleaning

```python
import pandas as pd
import datetime

def convert_to_datetime(date_value):
    if isinstance(date_value, str):
        if '/' in date_value:
            return pd.Timestamp(pd.to_datetime(date_value, format='%m/%d/%Y')).date()
        elif '-' in date_value:
            return pd.Timestamp(pd.to_datetime(date_value, format='%Y-%m-%d')).date()
        else:
            raise ValueError("Unsupported date format")
    elif isinstance(date_value, datetime.datetime ):
        return date_value.date()
    elif isinstance(date_value, (int, float)):
        return pd.Timestamp(pd.to_datetime('1899-12-30') + pd.Timedelta(days=int(date_value) - 2)).date()
    else:
        raise ValueError("Unsupported date format")

data = {
    'Name': ['John', 'Jane', 'Jim', 'Joan'],
    'date_of_assessment': ['2022-01-01', 42000, datetime.datetime(2022, 1, 1), '02/01/2022']
}

df = pd.DataFrame(data)

for i, date_value in enumerate(df['date_of_assessment']):
    try:
        df.at[i, 'date'] = convert_to_datetime(date_value)
    except ValueError as e:
        print(f"Error converting date value {date_value}: {e}")

print(df)
```
