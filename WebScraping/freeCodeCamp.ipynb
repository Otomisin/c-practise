### Scrape and output in table format

```python
from bs4 import BeautifulSoup

with open('home.html', 'r') as html_file:
    content = html_file.read()
    soup = BeautifulSoup(content, 'html.parser')
    course_cards = soup.find_all('div', class_='card')

    # Print table header
    print("{:<30} {:<10} {:<50}".format('Course Name', 'Price', 'Description'))
    print("-" * 90)  # Print a simple line separator

    for course in course_cards:
        course_name = course.h5.text.strip()
        course_price = course.a.text.split()[-1]
        course_description = course.p.text.strip()

        # Print each row of the table
        print("{:<30} {:<10} {:<50}".format(course_name, course_price, course_description))
```

### Scrape and output in pandas table format

```python
import pandas as pd
from bs4 import BeautifulSoup

# Read the HTML content
with open('home.html', 'r') as html_file:
    content = html_file.read()

soup = BeautifulSoup(content, 'html.parser')
course_cards = soup.find_all('div', class_='card')

# Lists to store scraped data
course_names = []
course_prices = []
course_descriptions = []

# Extract data and append to lists
for course in course_cards:
    course_names.append(course.h5.text.strip())
    course_prices.append(course.a.text.split()[-1])
    course_descriptions.append(course.p.text.strip())

# Create a DataFrame
df_courses = pd.DataFrame({
    'Course Name': course_names,
    'Price': course_prices,
    'Description': course_descriptions
})

# Display the DataFrame
df_courses.head()
```

## MonsterJob

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch the web page
response = requests.get('https://www.monster.co.uk/jobs/q-it-jobs?so=p.h.p')

# List to hold job data
jobs_data = []


# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')
# Find all job cards
jobs = soup.find_all('a', class_='cached-bot-shared-skeletonstyle__CachedSkeletonCard-sc-15ar3p6-1 gtOvnP')

for job in jobs:
    # Extract job information
    title = job.find('div', class_='cached-bot-shared-skeletonstyle__CachedSkeletonCardHeaderTitle-sc-15ar3p6-3 fIJwsb').text
    company = job.find('h2', class_='cached-bot-sharedstyle__JobCardCompany-sc-13tnr74-1 fRXVQe').text
    salary = job.find('p', class_='cached-bot-sharedstyle__JobCardDetails-sc-13tnr74-3 fPLwxY').text
    location = job.find_all('p', class_='cached-bot-sharedstyle__JobCardDetails-sc-13tnr74-3 fPLwxY')[1].text
    short_description = job.find('p', class_='cached-bot-sharedstyle__JobCardShortDescription-sc-13tnr74-4 cTAhxt').text
    link = job['href']

    # Add job information to the list
    jobs_data.append({
        'Title': title,
        'Company': company,
        'Salary': salary,
        'Location': location,
        'Description': short_description,
        'Link': link
    })

# Create a DataFrame from the jobs data
jobs_df = pd.DataFrame(jobs_data)

# Display the DataFrame
# print(jobs_df.head(10))
# print(jobs_df.head())

jobs_df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Correct URL to request
url = 'https://dtm.iom.int/reports?search=&sort_by=field_published_date&sort_order=DESC'

# Correctly make the request and store the response
response = requests.get(url)

dtm_soup = BeautifulSoup(response.content, 'html.parser')

# Find all relevant containers that include both the report link and the date
# This might need adjustment based on the actual page structure
report_containers = dtm_soup.find_all('div', class_='views-row')  # Example class, adjust based on actual HTML

# Initialize a list to hold tuples of (title, link, date)
report_details = []

# Loop through all found containers
for container in report_containers:
    # Find the 'a' tag within the container
    a_tag = container.find('a', href=True)
    # Check if 'href' exists and contains '/reports/' but not ending with '.pdf'
    if a_tag and '/reports/' in a_tag['href'] and not a_tag['href'].endswith('.pdf'):
        # Form the complete link
        link = f"https://dtm.iom.int{a_tag['href']}"
        # Extract the title (text content) of the 'a' tag
        title = a_tag.get_text(strip=True)
        # Attempt to find the date within the container
        date_div = container.find('div', class_='date')
        if date_div:
            date = date_div.text.split('·')[0].strip()  # Extract the published date
        else:
            date = "Date not found"  # Placeholder if date is not found
        
        # Append the tuple (title, link, date) to our list if the title is not empty
        if title:  # This checks if the title is not just whitespace
            report_details.append((title, link, date))

# Creating a DataFrame from the list of tuples
df_reports = pd.DataFrame(report_details, columns=['Title', 'Link', 'Date'])

# Display the DataFrame
print(df_reports)
```
