### Scrape and output in table format

```python
from bs4 import BeautifulSoup

with open('home.html', 'r') as html_file:
    content = html_file.read()
    soup = BeautifulSoup(content, 'html.parser')
    course_cards = soup.find_all('div', class_='card')

    # Print table header
    print("{:<30} {:<10} {:<50}".format('Course Name', 'Price', 'Description'))
    print("-" * 90)  # Print a simple line separator

    for course in course_cards:
        course_name = course.h5.text.strip()
        course_price = course.a.text.split()[-1]
        course_description = course.p.text.strip()

        # Print each row of the table
        print("{:<30} {:<10} {:<50}".format(course_name, course_price, course_description))
```

### Scrape and output in pandas table format

```python
import pandas as pd
from bs4 import BeautifulSoup

# Read the HTML content
with open('home.html', 'r') as html_file:
    content = html_file.read()

soup = BeautifulSoup(content, 'html.parser')
course_cards = soup.find_all('div', class_='card')

# Lists to store scraped data
course_names = []
course_prices = []
course_descriptions = []

# Extract data and append to lists
for course in course_cards:
    course_names.append(course.h5.text.strip())
    course_prices.append(course.a.text.split()[-1])
    course_descriptions.append(course.p.text.strip())

# Create a DataFrame
df_courses = pd.DataFrame({
    'Course Name': course_names,
    'Price': course_prices,
    'Description': course_descriptions
})

# Display the DataFrame
df_courses.head()
```

## MonsterJob

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch the web page
response = requests.get('https://www.monster.co.uk/jobs/q-it-jobs?so=p.h.p')

# List to hold job data
jobs_data = []


# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')
# Find all job cards
jobs = soup.find_all('a', class_='cached-bot-shared-skeletonstyle__CachedSkeletonCard-sc-15ar3p6-1 gtOvnP')

for job in jobs:
    # Extract job information
    title = job.find('div', class_='cached-bot-shared-skeletonstyle__CachedSkeletonCardHeaderTitle-sc-15ar3p6-3 fIJwsb').text
    company = job.find('h2', class_='cached-bot-sharedstyle__JobCardCompany-sc-13tnr74-1 fRXVQe').text
    salary = job.find('p', class_='cached-bot-sharedstyle__JobCardDetails-sc-13tnr74-3 fPLwxY').text
    location = job.find_all('p', class_='cached-bot-sharedstyle__JobCardDetails-sc-13tnr74-3 fPLwxY')[1].text
    short_description = job.find('p', class_='cached-bot-sharedstyle__JobCardShortDescription-sc-13tnr74-4 cTAhxt').text
    link = job['href']

    # Add job information to the list
    jobs_data.append({
        'Title': title,
        'Company': company,
        'Salary': salary,
        'Location': location,
        'Description': short_description,
        'Link': link
    })

# Create a DataFrame from the jobs data
jobs_df = pd.DataFrame(jobs_data)

# Display the DataFrame
# print(jobs_df.head(10))
# print(jobs_df.head())

jobs_df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Set the URL to request
url = 'https://www.theguardian.com/world/migration'

# Headers to simulate a browser visit
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Make the request with headers
response = requests.get(url, headers=headers)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.content, 'html.parser')

# Initialize a list to hold the extracted data
data = []

# Extract information from the specific parts of the page
for section in soup.find_all('section', class_='fc-container'):
    date_text = section.find('time', class_='fc-date-headline').text.strip()
    articles = section.find_all('div', class_='fc-item__container')
    for article in articles:
        title_tag = article.find('a', class_='fc-item__link')
        link = title_tag['href'] if title_tag else 'No link available'
        title = title_tag.text.strip() if title_tag else 'No title available'
        timestamp_tag = article.find('time', class_='fc-item__timestamp')
        published_time = timestamp_tag['datetime'] if timestamp_tag else 'Timestamp not available'
        
        # Append the extracted information to the list
        data.append({
            'Date': date_text,
            'Link': link,
            'Title': title,
            'Published Time': published_time
        })

# Create a DataFrame from the extracted data
data_df = pd.DataFrame(data)

# Display the DataFrame
data_df.head(10)
```

### Ajazera

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Set the URL to request
url = 'https://www.theguardian.com/world/migration'

# Headers to simulate a browser visit
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Make the request with headers
response = requests.get(url, headers=headers)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.content, 'html.parser')

# Initialize a list to hold the extracted data
data = []

# Extract information from the specific parts of the page
for article in soup.find_all('article', class_='gc'):
    # Extracting the title and link
    title_tag = article.find('a', class_='u-clickable-card__link')
    title = title_tag.text.strip() if title_tag else 'No title available'
    link = title_tag['href'] if title_tag else 'No link available'
    
    # Extracting the publication date
    date_tag = article.find('span', class_='screen-reader-text')
    date_text = date_tag.text.replace('Published On ', '').strip() if date_tag else 'Date not available'
    
    # Extracting the description
    description_tag = article.find('div', class_='gc__excerpt')
    description = description_tag.text.strip() if description_tag else 'No description available'
    
    # Append the extracted information to the list
    data.append({
        'Title': title,
        'Link': 'https://www.theguardian.com' + link,
        'Published Date': date_text,
        'Description': description
    })

# Create a DataFrame from the extracted data
data_df = pd.DataFrame(data)

# Display the DataFrame
data_df.head(10)
```

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

root = "https://www.google.com/"
link = "https://www.google.com/search?q=biden&tbm=nws&source=lnt&tbs=sbd:1&sa=X&ved=0ahUKEwjAvsKDyOXtAhXBhOAKHXWdDgcQpwUIKQ&biw=1604&bih=760&dpr=1.2"

def news(link):
    req = Request(link, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urlopen(req).read()
    soup = BeautifulSoup(webpage, 'html5lib')
    for item in soup.find_all('div', attrs={'class': 'ZINbbc xpd O9g5cc uUPGi'}):
        raw_link = (item.find('a', href=True)['href'])
        link = (raw_link.split("/url?q=")[1]).split('&sa=U&')[0]
                
        title = (item.find('div', attrs={'class': 'BNeawe vvjwJb AP7Wnd'}).get_text())
        description = (item.find('div', attrs={'class': 'BNeawe s3v9rd AP7Wnd'}).get_text())

        title = title.replace(",", "")
        description = description.replace(",", "")

        time = description.split(" · ")[0]
        descript = description.split(" · ")[1]
        
        print(title)
        print(time)
        print(descript)
        print(link)
        document = open("Data.csv", "a")
        document.write("{}, {}, {}, {} \n".format(title, time, descript, link))
        document.close()
        
    next = soup.find('a', attrs={'aria-label':'Next page'})
    next = (next['href'])
    link = root + next
    news(link)
news(link)
```

```python
from pygooglenews import GoogleNews
gn = GoogleNews(country = 'UK')

def get_titles(search):
    stories = []
    search = gn.search(search)
    newsitem = search['entries']
    for item in newsitem:
        story = {
            'title': item.title,
            'link': item.link
        }
        stories.append(story)
    return stories

print(get_titles('lockdown'))
```

```python
import pandas as pd
from gnews import GNews

# Create an instance of GNews
google_news = GNews()

# Get news articles about NASA
articles = google_news.get_news('NASA')

# Initialize lists to store data
dates = []
sources = []
links = []
titles = []
summaries = []

# Extract data from each article
for article in articles:
    dates.append(article['date'])
    sources.append(article['source'])
    links.append(article['link'])
    titles.append(article['title'])
    summaries.append(article['desc'])

# Create a DataFrame
df = pd.DataFrame({
    'Source': sources,
    'Link': links,
    'Title': titles,
    'Summary': summaries
})

# Display the DataFrame
print(df)

```
