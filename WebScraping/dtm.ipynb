```python
import requests
from bs4 import BeautifulSoup

# The URL of the site you want to scrape
url = "https://dtm.iom.int/reports"

# Fetch the content of the webpage
response = requests.get(url)
if response.status_code == 200:
    # Use BeautifulSoup to parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all elements that contain the report titles, assuming they are in <a> tags under a specific class
    # You might need to inspect the web page to find the correct tags and classes for report titles
    report_title_elements = soup.find_all('a', class_='content') # Note: Replace 'report-title-class' with the actual class name

    # Extract and print the text from each element
    for title_element in report_title_elements:
        print(title_element.text.strip())
else:
    print("Failed to retrieve the webpage")

```
