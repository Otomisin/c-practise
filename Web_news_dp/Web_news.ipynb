### theGuardian

- It has pages

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Set the URL to request
url = 'https://www.theguardian.com/world/migration'

# Headers to simulate a browser visit
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Make the request with headers
response = requests.get(url, headers=headers)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.content, 'html.parser')

# Initialize a list to hold the extracted data
data = []

# Extract information from the specific parts of the page
for section in soup.find_all('section', class_='fc-container'):
    date_text = section.find('time', class_='fc-date-headline').text.strip()
    articles = section.find_all('div', class_='fc-item__container')
    for article in articles:
        title_tag = article.find('a', class_='fc-item__link')
        link = title_tag['href'] if title_tag else 'No link available'
        title = title_tag.text.strip() if title_tag else 'No title available'
        timestamp_tag = article.find('time', class_='fc-item__timestamp')
        published_time = timestamp_tag['datetime'] if timestamp_tag else 'Timestamp not available'
        
        # Append the extracted information to the list
        data.append({
            'Date': date_text,
            'Link': link,
            'Title': title,
            'Published Time': published_time
        })

# Create a DataFrame from the extracted data
data_df = pd.DataFrame(data)

# Display the DataFrame
data_df.head(10)
```

### Ajazera

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Set the URL to request
url = 'https://www.theguardian.com/world/migration'

# Headers to simulate a browser visit
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Make the request with headers
response = requests.get(url, headers=headers)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.content, 'html.parser')

# Initialize a list to hold the extracted data
data = []

# Extract information from the specific parts of the page
for article in soup.find_all('article', class_='gc'):
    # Extracting the title and link
    title_tag = article.find('a', class_='u-clickable-card__link')
    title = title_tag.text.strip() if title_tag else 'No title available'
    link = title_tag['href'] if title_tag else 'No link available'
    
    # Extracting the publication date
    date_tag = article.find('span', class_='screen-reader-text')
    date_text = date_tag.text.replace('Published On ', '').strip() if date_tag else 'Date not available'
    
    # Extracting the description
    description_tag = article.find('div', class_='gc__excerpt')
    description = description_tag.text.strip() if description_tag else 'No description available'
    
    # Append the extracted information to the list
    data.append({
        'Title': title,
        'Link': 'https://www.theguardian.com' + link,
        'Published Date': date_text,
        'Description': description
    })

# Create a DataFrame from the extracted data
data_df = pd.DataFrame(data)

# Display the DataFrame
data_df.head(10)
```

```python
from pygooglenews import GoogleNews
gn = GoogleNews(country = 'UK')

def get_titles(search):
    stories = []
    search = gn.search(search)
    newsitem = search['entries']
    for item in newsitem:
        story = {
            'title': item.title,
            'link': item.link
        }
        stories.append(story)
    return stories

print(get_titles('lockdown'))
```

Generating the countries using the country_data.son file

```python
import pandas as pd
import re
import json

# Load country data from the uploaded JSON file
with open('country_data.json', 'r', encoding='utf-8') as file:
    country_data = json.load(file)

# Generate a regex pattern for each country using preferred names and regex from the data
country_patterns = {}
for country in country_data:
    preferred_name = country["#country+name+preferred"]
    regex = country["#country+regex"]
    # Use the regex if it's not empty; otherwise, escape the preferred name and use it as regex
    pattern = re.escape(preferred_name) if not regex else regex
    country_patterns[preferred_name] = re.compile(pattern, re.IGNORECASE)


df_new = df

# Function to extract country names or return "No Country Names"
def extract_countries(text):
    countries = set()
    for country, pattern in country_patterns.items():
        if pattern.search(text):
            countries.add(country)
    return ', '.join(countries) if countries else 'No Country Names'

# Apply the function to 'title' and 'description' columns and handle duplicates
df_new['Country'] = df_new.apply(lambda row: extract_countries(row['title']) + ', ' + extract_countries(row['description']), axis=1)
df_new['Country'] = df_new['Country'].apply(lambda x: ', '.join(set(x.split(', '))))  # Remove duplicates

# Display the DataFrame
df_new
```
