```python
import pyspark
pyspark.SparkContext.setSystemProperty('spark.ui.showConsoleProgress', 'true')
pyspark.SparkContext.setSystemProperty('spark.executor.extraJavaOptions', '-verbose:class')
pyspark.SparkContext.setSystemProperty('spark.driver.extraJavaOptions', '-verbose:class')

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Test App").getOrCreate()
df = spark.createDataFrame([("John", 1000), ("Jane", 2000)], ["Name", "Salary"])
df.show()
spark.stop()
```
